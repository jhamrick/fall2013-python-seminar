# Efficiency and Purity Plots

Adapted from Morgan et al. 2012, *The Astrophysical Journal*, 746, 170

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define data type for purity and efficiency data sets
dtype = np.dtype(
       [("frac_followed", "f8"), 
        ("frac_observed", "f8"), 
        ("frac_observed_uncertainty", "f8")])
```

```python
# load efficiency data set
efficiency = np.loadtxt(
    "hw_3_1_example/Efficiency.txt", 
    skiprows=1, 
    dtype=dtype)
```

```python
# load purity data set
purity = np.loadtxt(
    "hw_3_1_example/Purity.txt",
    skiprows=2,
    dtype=dtype)
```

```python
# properties for annotation arrows
arrowprops = {
    'arrowstyle': '->',
    'linewidth': 2
}
```

```python
# create the figure and two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True)

# plot a solid diagonal line
ax1.plot([0, 1.0], [0, 1.0], color='k', lw=1)

# plot the efficiency data
ax1.fill_between(
    efficiency["frac_followed"],
    efficiency["frac_observed"] - efficiency["frac_observed_uncertainty"],
    efficiency["frac_observed"] + efficiency["frac_observed_uncertainty"],
    color="k", alpha=0.3)
ax1.plot(efficiency["frac_followed"], efficiency["frac_observed"],
    color="k")

# plot a vertical dotted line for high-z events
idx, = np.nonzero(np.round(efficiency["frac_followed"], decimals=2) == 0.2)
yval = efficiency["frac_observed"][idx[0]]
ax1.plot([0.2, 0.2], [-0.05, yval], 'k:', lw=2)

# annotate the high-z events
ax1.annotate(
    "Follow up 20% of\nbursts to capture ~55%\nof high-z events",
    [0.2, yval], [0.35, 0.1], 
    arrowprops=arrowprops)

# annotate random guessing
ax1.annotate(
    "Random\nguessing",
    [0.8, 0.8], [0.8, 0.5], 
    arrowprops=arrowprops, 
    horizontalalignment='center')

# set axis labels and title
ax1.set_xlabel("Fraction of GBRs Followed Up")
ax1.set_ylabel("Fraction of high (z > 4) GBRs observed")
ax1.set_title("Efficiency", fontsize=16, fontweight="bold")

# plot a solid horizontal line at the minimum observed for purity
ax2.plot([0, 1.0],[purity["frac_observed"].min()]*2, color='k', lw=1)

# plot the purity data
ax2.fill_between(
    purity["frac_followed"],
    purity["frac_observed"] - purity["frac_observed_uncertainty"],
    purity["frac_observed"] + purity["frac_observed_uncertainty"],
    color="k", alpha=0.3)
ax2.plot(purity["frac_followed"], purity["frac_observed"],
    color="k")

# plot a vertical dotted line for followed up events
idx, = np.nonzero(np.round(purity["frac_followed"], decimals=2) == 0.2)
yval = purity["frac_observed"][idx[0]]
ax2.plot([0.2, 0.2], [-0.05, yval], 'k:', lw=2)

# annotate followed up events
ax2.annotate(
    "If 20% of events are\nfollowed up, ~40% of\nthem will be high-z",
    [0.2, yval], [0, 0.8], 
    arrowprops=arrowprops)

# annotate random guessing
ax2.annotate(
    "Random\nguessing",
    [0.6, purity["frac_observed"].min()], [0.7, 0.5], 
    arrowprops=arrowprops)

# set axis labels and title
ax2.set_xlabel("Fraction of GBRs Followed Up")
ax2.set_ylabel("Fraction of observed GBRs that are high z (z > 4)")
ax2.set_title("Purity", fontsize=16, fontweight="bold")

# customize axis limits and ticks
for ax in [ax1, ax2]:
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    ax.tick_params(direction='out')
    ax.set_xlim(-0.05, 1.05)
    ax.set_ylim(-0.05, 1.05)

# set the figure size
fig.set_figwidth(9)
fig.set_figheight(4)

# add a little more space between the plots
plt.subplots_adjust(wspace=0.3)
```

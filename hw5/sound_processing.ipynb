```python
# process all the sound files, saving a plot of the scale vs. power
# and reporting the identified notes
if not os.path.exists("notes"):
    os.makedirs("notes")
for filename in files:
    name = os.path.splitext(os.path.split(filename)[1])[0]
    # load the file
    data, rate = sound.load(filename)
    # compute frequency and power
    freq, power = sound.freq_power(data, rate)
    # identify the notes from the power spectrum
    bins, notes, Y = sound.bin_notes(freq, power)
    notes = sound.pick_notes(Y)
    # plot the notes and save the figure
    sound.plot_notes(bins, Y)
    plt.title("%s (%s)" % (name, "+".join(notes)))
    plt.savefig("notes/%s.png" % name)
    plt.close('all')
    
    print filename, "+".join(notes)
```

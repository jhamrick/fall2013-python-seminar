Write an XML-RPC server/client program pair.

* The server offers 3 methods providing lossless manipulation (be
  creative). Each server method must have descriptive documentation
  accessible by the client.

* The client calls each method on the server machine with an array

* Client saves the original image and modified image received from the

```python
from xmlrpclib import ServerProxy
from xmlimage import encode, decode
import matplotlib.pyplot as plt
```

```python
image = plt.imread("panda.jpg")
plt.imshow(image)
```

```python
host, port = "127.0.0.1", 5021
server = ServerProxy("http://%s:%d" % (host, port))
```

```python
available_methods = server.system.listMethods()
print "Available methods from server:"
for method in available_methods:
    print "\t" + method
```

```python
def call(func, image):
    return encode(func(decode(image)))
```

```python
plt.imshow(call(server.invert, image))
```

```python
plt.imshow(call(server.flip_horizontal, image))
```

```python
plt.imshow(call(server.flip_vertical, image))
```

```python
plt.imshow(call(server.rotate_clockwise, image))
```

```python
plt.imshow(call(server.rotate_counterclockwise, image))
```

Write an XML-RPC server/client program pair.

* The server offers 3 methods providing lossless manipulation (be

  accessible by the client.

* The client calls each method on the server machine with an array

* Client saves the original image and modified image received from the

```python
%matplotlib inline
```

```python
import xmlrpclib
import sys
import matplotlib.pyplot as plt
import xmlimage
```

```python
image = plt.imread("panda.jpg")
plt.imshow(image)
```

```python
host, port = "127.0.0.1", 5021
server = xmlrpclib.ServerProxy("http://%s:%d" % (host, port))
```

```python
available_methods = server.system.listMethods()
print "Available methods from server:"
for method in available_methods:
    print "\t" + method
```

```python
plt.imshow(xmlimage.encode(server.invert(xmlimage.decode(image))))
```

```python
plt.imshow(xmlimage.encode(server.flip_horizontal(xmlimage.decode(image))))
```

```python
plt.imshow(xmlimage.encode(server.flip_vertical(xmlimage.decode(image))))
```

```python
plt.imshow(xmlimage.encode(server.rotate_clockwise(xmlimage.decode(image))))
```

```python
plt.imshow(xmlimage.encode(server.rotate_counterclockwise(xmlimage.decode(image))))
```
